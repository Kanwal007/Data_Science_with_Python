## Plotting in Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter plot
# Change Colors
palette = {"male": "blue", "female": "pink"}
plt.figure(figsize=(10,8)) 
sns.scatterplot(data=df, x='age', y='fare',hue='sex',palette=palette)
```

```python
plt.savefig('scatterplot.png',)
```

# Interpretations
1. we have age at x-axis and fare at y-axis
2. Scatter plot shows the relationship between age and fare of titanic passengers
3. we further grouped the passenger by gender
4. Two male passengers gave the highest fare in titanic, and they might have travelled first class
5. Oldest person is also a male having 80 years of age

```python
sns.boxplot(data=df, x='sex', y='fare',hue='class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
```

```python
import plotly.express as px
fig = px.scatter(df, x="age", y="fare", color="sex", facet_col="class")
fig.show()
```
